# Machine Learning
Consiste à donner à une machine la capacité d'apprendre sans l'a programmée de facon explicite. 
## L'apprenditssage supervise 
Consiste à montrer à la machine des examples X Y y le demander en fait une association y égale  Y=F(X)
### Des notions foundamentales
1. **Le Dataset** Target variable. C'est que la machine apprendre a predir. Avec deux types de variables, 
Target Y (m: nombre d'examples), 
Features x1, x2, x3 (n: nombre de features)
Le Dataset contient les questions (x) et les réponses (y) au probléme que la machine doit résoudre.


2. **Modèle**. Paramètres,  on crée un modèle, qui n'est autre qu'une fontion mathématique, les coefficientes du polynôme abcd autant de coefficient qu'on a.
C'est nous qui décidons quel modèles doit utiliser et la machine doit aprrendre les paramètres de ce modèle.  
3. **Fontion coût**. Il donne des erreurs. Avoir un bon modèle c'est avoir un modèle qui nous donnent de petites erreurs.
4. **Algorithme de minimisation**. Développer une stratégie qui cherche à trouver quels sont les paramètres qui minimisent la funtion coût, on utilise un algorithme d'apprentissage, par example l'algorithme grayden descent.  

## Les applications du Supervised Learning

On peut développer des modèles pour résoudre 2 types de problèmes: 
* Les problèmes de **Régression**: on cherche à predire la valeur d'une variable continue, c'est-à-dire une varible qui peut prendre une infinité de valeurs. 
    Par example: 
    * Prédire le prix d'un appartement (y) selon sa surface habitable (x)
    * Prédire la quantite d'essence consommée (y) selon la distance parcourue(x)
* Les problèmes de **Classification**: on cherche à classer un objet dans différentes classes, c'est-à-dire que l'on cherche à predir la valeur d'une variable discrète qui prend un nombre fini de valeurs.
    Par example: 
    * Prédire si un email est un spam (clase y=1) ou non (classe y= 0) selon le nombre de liens présent dans l'email (x)
    * Prédire si une tumeur est maligne (y=1) ou bénigne (y=0) selon la taille de la tumeur (x1) et l'âge du patient (x2)
