## Les applications du Supervised Learning

On peut dvelopper des modles pour rsoudre 2 types de problmes: 
* Les problmes de **Rgression**: on cherche  predire la valeur d'une variable continue, c'est--dire une varible qui peut prendre une infinit de valeurs. 
    Par example: 
    * Prdire le prix d'un appartement (y) selon sa surface habitable (x)
    * Prdire la quantite d'essence consomme (y) selon la distance parcourue(x)
* Les problmes de **Classification**: on cherche  classer un objet dans diffrentes classes, c'est--dire que l'on cherche  predir la valeur d'une variable discrte qui prend un nombre fini de valeurs.
    Par example: 
    * Prdire si un email est un spam (clase y=1) ou non (classe y= 0) selon le nombre de liens prsent dans l'email (x)
    * Prdire si une tumeur est maligne (y=1) ou bnigne (y=0) selon la taille de la tumeur (x1) et l'ge du patient (x2)
