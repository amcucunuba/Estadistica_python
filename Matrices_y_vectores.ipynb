# Algebra lineal 

## Espacio vectorial 

En álgebra lineal, un espacio vectorial (o también llamado espacio lineal) es una **estructura algebraica creada a partir de un conjunto no vacío** 
* Una operación interna (llamada suma, definida para los elementos del conjunto) (grupo abeliano)
* Una operación externa (llamada producto por un escalar, definida entre dicho conjunto y otro conjunto, con estructura de cuerpo) que satisface 8 propiedades fundamentales. (cuerpo, numeros racionales/reales, complejos)

A **los elementos de un espacio vectorial** se les llama **vectores** y a los **elementos del cuerpo** se les conoce **como escalares**

En programación, las estructuras de datos (arrays o vectores, matrices…) es la forma que tenemos para poder **organizar, almacenar y hacer uso de los datos de la forma más eficiente** y nos van a dar más opciones para resolver programas más complejos

### Una matriz 
Es un **conjunto** de **numeros ordenados** en forma de **filas y columnas**.

Las matrices se **usan** generalmente para **describir sistemas de ecuaciones lineales**, **sistemas de ecuaciones diferenciales** o **representar una aplicación lineal**

Para sumar o restar matrices, estas deben tener las mismas dimensiones, es decir, el mismo número de filas y columnas.

La multiplicación de matrices es posible cuando el número de columnas de la primera matriz coincide con el número de filas de la segunda. El resultado es una nueva matriz cuyas dimensiones vienen dadas por las filas de la primera y las columnas de la segunda matriz.

Las matrices son un **array bidimensional**

```python
#Calcular 10 número de manera aleatoria
import random

def vector_aleatorio(n):
    vector= [0]*n 
    for i in range(n):
        vector[i] = random.randint(0, 10)
        '''El valor debe estar definido, no puede ser infinito 
        como si lo hacen las lista'''
    return vector

print ('Ingrese cuantos numeros aleatorios desea')
n = int(input())

aleatorio = vector_aleatorio(n)
print (aleatorio)
```
